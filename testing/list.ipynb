```python
import numpy as np
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
class urlcont:
    def __init__(self,url):
        self.url=url
        self.r = requests.get(url) 
        self.soup = bs(self.r.content, 'html5lib')
        
    def count_comments(self):
        ans=[]
        out=""
        comment=0
        like=0
        dislike=0
        cou=self.soup.find_all('iframe')
        sp=self.url.split("/")
        if sp[2]=="www.ndtv.com":
            iframe=cou[len(cou)-1]
        elif sp[2]=="sports.ndtv.com":
            iframe=cou[0]
        request=requests.get(iframe.attrs['src'])
        iframe_soup=bs(request.content,'html5lib')
        anc=iframe_soup.find_all('a')
        for an in anc:
            text=an.get_text()
            text=text.strip()
            out=text
            if "Like"==text:
                comment+=1
            elif "Unlike"==text:
                out=text
            elif "Like" in text:
                comment+=1
                w=text.split(" ")
                e=w[1]
                e=e.split("(")
                e="".join(e)
                e=e.split(")")
                e="".join(e)
                like=like+int(e)
            elif "Unlike" in text:
                w=text.split(" ")
                e=w[1]
                e=e.split("(")
                e="".join(e)
                e=e.split(")")
                e="".join(e)
                dislike=dislike+int(e)
        ans.append(comment)
        ans.append(like)
        ans.append(dislike)
        if out=="More Comments":
            ans.append(1) 
        else:
            ans.append(0)
        return ans
            
        
```

```python
fname = 'primary.html' #this html file is stored on the same folder of the code file
html_file = open(fname, 'r',encoding='utf-8')
code = html_file.read() 
```

```python
r=bs(code,'html5lib')
```

```python
urls=r.find('ul',id='ui-id-62')
```

```python
anc=urls.find_all('a')
```

```python
final_ans=[]
inner=[]
for href in anc:
    url=href.attrs['href']
    sp=url.split("/")
    if (sp[2]=="www.ndtv.com")or sp[2]=="sports.ndtv.com":
        if sp[3]!="tamil" and sp[3]!="bengali" and sp[3]!="hindi" and url!="https://www.ndtv.com/india-news/bjp-lok-sabha-election-results-2019-live-news-updates-seats-will-bjp-return-to-power-2041503":
            a=urlcont(url)
            print(url)
            b=a.count_comments()
            inner=[]
            inner.append(30)
            inner.append(url)
            inner.append(b[0])
            inner.append(b[1])
            inner.append(b[2])
            inner.append(b[3])
            final_ans.append(inner)
```

```python
import pandas as pd
data=pd.DataFrame(final_ans,columns=['Date','URL','comments','likes','dislikes','recheck'])
```

```python
data.head(10)
```

```python
data1=data[data['comments']>=10]
data2=data[(data.comments >= 5) & (data.comments <10)]
data3=data[(data.comments >0) & (data.comments <5)]
data4=data[data['comments']==0]

data1=data1.sort_values('comments',ascending=False)
data2=data2.sort_values('comments',ascending=False)
data3=data3.sort_values('comments',ascending=False)
data4=data4.sort_values('comments',ascending=False)
```

```python
samp1=data1.sample(n=6)
samp2=data2.sample(n=6)
samp3=data3.sample(n=6)
samp4=data4.sample(n=2)
```

```python
samp=pd.concat([samp1,samp2,samp3,samp4],axis=0)
samp.shape
output=samp.values
output
```
